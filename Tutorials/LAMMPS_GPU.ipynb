## Install LAMMPS package to use in GPU

```python
# Install Prerequisites
!apt install cmake build-essential git ccache

# Dowload LAMMPS
!rm -rf lammps
!git clone -b stable --depth=1 https://github.com/lammps/lammps.git lammps

# Build LAMMPS with GPU and PYTHON modules (Add more if needed with PKG_NAME=on options)
!rm -rf build
!mkdir build
!cd lammps/cmake/presets; cat basic.cmake | sed "s/RIGID/RIGID GPU EXTRA-PAIR/" > kk ; mv kk basic.cmake
!cd build; cmake -C ../lammps/cmake/presets/basic.cmake -D CMAKE_INSTALL_PREFIX=/usr -D CMAKE_CXX_COMPILER_LAUNCHER=ccache -D BUILD_SHARED_LIBS=on -D LAMMPS_EXCEPTIONS=on -D PKG_PYTHON=on -D PKG_GPU=on -D GPU_API=cuda -D GPU_PREC=mixed -D GPU_ARCH=sm_75 ../lammps/cmake
!cd build; make -j 2

# Compile and Install LAMMPS
!cd build; make install
```
